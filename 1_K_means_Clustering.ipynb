```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(5)
X = np.random.randn(90,2)
plt.plot(X[:,0],X[:,1],"o")
plt.show()
```

```python
X
```

```python
#Generate three Clusters
X[:30,0] -= 5
X[:30,1] -= 5

X[30:60,0] += 5
X[30:60,1] += 5
```

```python
plt.plot(X[:,0],X[:,1],"o")
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters=3,random_state=42)
km.fit(X)
```

```python
km.cluster_centers_
```

```python
km.labels_
```

```python
plt.scatter(X[:,0],X[:,1],c=km.labels_,cmap = "viridis")
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],s = 250,
            marker = "*",
            c = "red",label = "Centroids")
plt.xlabel("X Axis")
plt.ylabel("X Axis")
plt.title("K Means")
plt.colorbar(label="Clusters")
plt.legend()
plt.show()
```

```python
p = [[0,2]]
km.predict(p)
```

```python
distortions = []
for i in range(1, 11):#10
    km = KMeans(n_clusters=i,
                init='k-means++',
                n_init=10,
                max_iter=300,
                random_state=0)
    km.fit(X)
    distortions.append(km.inertia_)
plt.plot(range(1, 11), distortions, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Distortion')
plt.show()
```

## Silhouette Score
- value that ranges between [-1,1]
- Silhouette Score of 1, means clusters are dense and they are nicely separated. When score is 0, it means clusters are overlapping. The score less than 0, means data belonging to clusters may be wrong

```python
#Silhouette Score
from sklearn.metrics import silhouette_score
```

```python
#Silhouette Method
from sklearn.metrics import silhouette_score
```

```python
kmeans_per_k = [KMeans(n_clusters=k, random_state=42).fit(X)
                for k in range(1, 10)]
silhouette_scores = [silhouette_score(X, model.labels_)
                     for model in kmeans_per_k[1:]]
print(silhouette_scores)
plt.figure(figsize=(8, 3))
plt.plot(range(2, 10), silhouette_scores, "bo-")
plt.xlabel("$k$", fontsize=14)
plt.ylabel("Silhouette score", fontsize=14)
plt.show()
```

```python
from sklearn.metrics import silhouette_samples
from matplotlib.ticker import FixedLocator, FixedFormatter
import matplotlib as mpl

plt.figure(figsize=(11, 9))

for k in (3, 4, 5, 6):
    plt.subplot(2, 2, k - 2)

    y_pred = kmeans_per_k[k - 1].labels_
    silhouette_coefficients = silhouette_samples(X, y_pred)

    padding = len(X) // 30
    pos = padding
    ticks = []
    for i in range(k):
        coeffs = silhouette_coefficients[y_pred == i]
        coeffs.sort()

        color = mpl.cm.Spectral(i / k)
        plt.fill_betweenx(np.arange(pos, pos + len(coeffs)), 0, coeffs,
                          facecolor=color, edgecolor=color, alpha=0.7)
        ticks.append(pos + len(coeffs) // 2)
        pos += len(coeffs) + padding

    plt.gca().yaxis.set_major_locator(FixedLocator(ticks))
    plt.gca().yaxis.set_major_formatter(FixedFormatter(range(k)))
    if k in (3, 5):
        plt.ylabel("Cluster")

    if k in (5, 6):
        plt.gca().set_xticks([-0.1, 0, 0.2, 0.4, 0.6, 0.8, 1])
        plt.xlabel("Silhouette Coefficient")
    else:
        plt.tick_params(labelbottom=False)

    plt.axvline(x=silhouette_scores[k - 2], color="red", linestyle="--")
    plt.title("$k={}$".format(k), fontsize=16)

#save_fig("silhouette_analysis_plot")
plt.show()
```
